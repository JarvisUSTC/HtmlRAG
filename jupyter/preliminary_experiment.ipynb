```python
import string
import re
import json

#  evaluate with short answer exact match
def normalize_answer(s):
  """Lower text and remove punctuation, articles and extra whitespace."""

  def remove_articles(text):
    return re.sub(r'\b(a|an|the)\b', ' ', text)

  def white_space_fix(text):
    return ' '.join(text.split())

  def remove_punc(text):
    exclude = set(string.punctuation)
    return ''.join(ch for ch in text if ch not in exclude)

  def lower(text):
    return text.lower()

  return white_space_fix(remove_articles(remove_punc(lower(s))))

def calculate_short_answer_EM(generated_answer, gold_answers, language="en"):
    if isinstance(gold_answers, str):
        gold_answers = [gold_answers]
    if language == "en":
        generated_answer=normalize_answer(generated_answer)
    else:
        pass
    match=0
    for gold_answer in gold_answers:
        gold_answer=normalize_answer(gold_answer)
        if gold_answer in generated_answer:
            match+=1
    return {
        "exact_match": match/len(gold_answers),
        "hit1": match>0
    }

dataset="asqa"
chat_model="bc34b192k"
split="test"
search_engine="bing"
multi_docs="top10"
rewrite_method="slimplmqr"
rerank_model="bgelargeen"

def eval_short_answer_EM(dataset, chat_model, reference_format, split, search_engine, rewrite_method, multi_docs):
    output_dir = f"../html_data/{dataset}/{chat_model}/{search_engine}"
    
    if dataset in ["wstqa", "statsgov", "sinafinance"]:
        language = "zh"
        # print(f"setting language to {language} for dataset {dataset}")
    elif dataset in ["websrc", "asqa", "hotpot-qa", "nq", "trivia-qa", "eli5", "musique"]:
        language = "en"
        # print(f"setting language to {language} for dataset {dataset}")
    else:
        raise NotImplementedError(f"dataset {dataset} not implemented")
    
    # output_file = f"{output_dir}/{chat_model}-{reference_format}-{dataset}-{split}.jsonl"
    # output_file = f"{output_dir}/{chat_model}-{reference_format}-{rewrite_method}-{multi_docs}-{dataset}-{split}.jsonl"
    # output_file = f"{output_dir}/{chat_model}-{reference_format}-{rewrite_method}-{rerank_model}-{dataset}-{split}.jsonl"
    output_file = f"{output_dir}/{chat_model}-{reference_format}-{rewrite_method}-{multi_docs}-{dataset}-{split}.jsonl"
    try:
        data_lines = [json.loads(l) for l in open(output_file)]
        if "answers" in data_lines[0]:
            answers = [data_line['answers'] for data_line in data_lines]
        elif "short_answers" in data_lines[0]:
            answers = [data_line['short_answers'] for data_line in data_lines]
        elif "answer" in data_lines[0]:
            answers = [data_line['answer'] for data_line in data_lines]
        else:
            raise NotImplementedError("answers not found in data_lines")
    
        generated_answers = [data_line[f"{chat_model}_{reference_format}" ] for data_line in data_lines]
        exact_match = [calculate_short_answer_EM(generated_answer, gold_answers) for generated_answer, gold_answers in zip(generated_answers, answers)]
        # print(f"chat_model: {chat_model}, reference_format: {reference_format}, dataset: {dataset}, split: {split}")
        # print({
        #     "hit1": round(sum([hit1["hit1"] for hit1 in exact_match])/len(exact_match)*100, 2),
        #     "exact_match": round(sum([hit1["exact_match"] for hit1 in exact_match])/len(exact_match)*100, 2)
        # })
        hit1= round(sum([hit1["hit1"] for hit1 in exact_match])/len(exact_match)*100, 2),
        exact_match= round(sum([hit1["exact_match"] for hit1 in exact_match])/len(exact_match)*100, 2)
        return {
            "hit1": hit1,
            "exact_match": exact_match
        }
    except Exception as e:
        print(f"error evaluating file {output_file}")
        #  print stack trace
        return {
            "hit1": .0,
            "exact_match": .0
        }
    
```

```python
import os
#  eval all datasets
#  generate markdown table report
context_windows=["2k", "4k", "8k", "16k", "32k", "64k",]
# context_windows=["64k"]
reference_formats=["raw_text", "markdown", "html-simple", "treegen", "html-trim"]
markdown_table = ["| dataset | ref | " + " | ".join(context_windows) + " |"]
markdown_table.append("| --- | --- | " + " | ".join(["---" for _ in range(len(context_windows))]) + " |")

datasets=["asqa", "hotpot-qa", "nq", "trivia-qa", "musique"]
# datasets=["hotpot-qa"]
    
split="test"
search_engine="bing"
multi_docs="top10"
rewrite_method="slimplmqr" 
version="v0715"


for dataset in datasets:
    for reference_format in reference_formats:
        markdown_table.append(f"| {dataset} | {reference_format} | ")

import concurrent.futures
with concurrent.futures.ProcessPoolExecutor(max_workers=16) as executor:
    for c, context_window in enumerate(context_windows):
        chat_model=f"llama70b{context_window}"
        
        for i, dataset in enumerate(datasets):
            # output_dir = f"../html_data/{dataset}/{chat_model}/{search_engine}"
            if dataset in ["wstqa", "statsgov", "sinafinance"]:
                language = "zh"
                # print(f"setting language to {language} for dataset {dataset}")
            elif dataset in ["websrc", "asqa", "hotpot-qa", "nq", "trivia-qa", "eli5", "musique"]:
                language = "en"
                # print(f"setting language to {language} for dataset {dataset}")
            else:
                raise NotImplementedError(f"dataset {dataset} not implemented")
        
            for j, reference_format in enumerate(reference_formats):
                if reference_format == "treegen":
                    future=executor.submit(eval_short_answer_EM, dataset, chat_model, reference_format, split, search_engine, rewrite_method, version)
                elif reference_format == "html-trim":
                    future=executor.submit(eval_short_answer_EM, dataset, chat_model, reference_format, split, search_engine, rewrite_method, rerank_model)
                else:
                    future=executor.submit(eval_short_answer_EM, dataset, chat_model, reference_format, split, search_engine, rewrite_method, multi_docs)
                res=future.result()["hit1"]
                if isinstance(res, tuple):
                    res=res[0]
                print(f"dataset: {dataset}, ref: {reference_format}, context_window: {context_window}, hit1: {res}")
                markdown_table[i*len(reference_formats)+j+2] += f"{res} |"   
    
markdown_str = "\n".join(markdown_table)
```

```python
import base64
from typing import Optional
import requests

#  retrieve nq bing search web pages
def parse_url(
        url: str,
        service_url: Optional[str] = None,
        need_links: bool = False,
        enable_render: bool = False,
        extract_level: int = 1,
) -> str:
    """
    Parse url into html text or raw text.
    Args:
        url (str): url to parse
        service_url (str, optional): service url. Defaults to None.
        enable_render (bool, optional): enable render on url to parse. Defaults to True.
        extract_level (int, optional): extract level. Defaults to 1. Here are the options:
            0: extract raw text only.  1: Just remove js, css, etc on url to parse.
            2: keep all. return raw html text. Note that 2, 3 will return base64 encoded html text.
    """
    if service_url is None:
        # service_url = "http://172.16.100.225:8081/fetch"  # online
        # service_url = "http://lb-2nshjbik-dfpxqkgc3sr5jomn.clb.ap-guangzhou.tencentclb.com/fetch"
        # service_url = "http://172.16.98.152:8081/fetch"  # offline
        service_url = "http://lb-3elxe8bu-3vp6fdo91l25hau7.clb.ap-guangzhou.tencentclb.com/fetch"
    payload = json.dumps({
        "url": url,
        "need_links": need_links,
        "enable_render": enable_render,
        "extract_level": extract_level,
    })
    print(f"parse url: {url}")
    headers = {'Content-Type': 'application/json'}
    response = requests.request("GET", service_url, headers=headers, data=payload)
    # print(f"response: {response.text}")
    if response.status_code != 200:
        raise Exception(f"Fail to request url: {url}, code: {response.status_code}")
    
    response_json = json.loads(response.text)
    # if response_json["status_code"] != 0:
    #     raise Exception(f"parse url {url} failed")
    parsed_html = response_json["content"]
    if extract_level != 0:
        try:
            parsed_html = base64.b64decode(parsed_html).decode("utf-8")
        except Exception:
            pass
    return parsed_html


import urllib3
from tqdm import tqdm
for idx in tqdm(range(len(data_lines[:2]))):
    #  send request to get html content
    for lidx, link in enumerate(data_lines[idx]['vanilla_search_results']):
        concat_url = link['url']
        # http = urllib3.PoolManager()
        # try:
        print(f"requesting {concat_url}")
        # response = http.request('GET', concat_url)
        response = parse_url(concat_url, extract_level=2)
        # except:
        #     print(f"error requesting {concat_url}")
        #     continue
        #  convert bytes to string
        # html = response.data.decode('utf-8')
        data_lines[idx]['vanilla_search_results'][lidx].update({
            "html": response
        })
```

```python
from urllib.parse import urlencode


class BingSearch(object):
    
    BASE_PATH = "https://api.bing.microsoft.com/v7.0/search?{}"
    SECRET_KEY = "d73c4b858f3f457a845598c4859e178b"
    
    @classmethod
    def _query(cls, query, count, mkt="zh-CN"):
        params = {
            "q": query,
            "textDecorations":False,
            "mkt":"en-US",
            "count": count,
            "SafeSearch": "Strict"
        }
        
        url = BingSearch.BASE_PATH.format(urlencode(params))
        # print(url)
        rsp = requests.get(url, headers={'Ocp-Apim-Subscription-Key': BingSearch.SECRET_KEY})
        if rsp is None or rsp.status_code != 200:
            return False, "http request failed"
        
        return True, rsp.json()
    
    @classmethod
    def search(cls, query: str, count=100, db_handler=None):
        ret, jdata = cls._query(query, count)
        if ret == False or 'webPages' not in jdata or 'value' not in jdata['webPages']:
            # logging.error("[search]-[error] query failed, query:{}, msg:{}".format(query, jdata))
            return []
        
        # print(json.dumps(jdata, ensure_ascii=False))
        return jdata['webPages']['value']
```
