```python
from bs4 import BeautifulSoup
from bs4.element import NavigableString

def build_document_tree(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')

    # 根节点
    root = {"tag": "root", "children": []}
    stack = [root]  # 栈，用于追踪当前层级的节点

    # 遍历所有子节点
    for tag in soup.descendants:
        if tag.name in ["h1", "h2", "h3", "h4", "h5", "h6"]:  # 层级节点
            level = int(tag.name[1])  # 获取层级数字
            node = {"tag": tag.name, "text": tag.get_text(), "children": []}

            # 确保栈中只有比当前标题层级高或等于的节点
            while len(stack) > level:
                stack.pop()

            # 插入到栈顶节点的 children 中
            stack[-1]["children"].append(node)

            # 将当前节点推入栈
            stack.append(node)
          
        elif isinstance(tag, NavigableString):
            # 文本节点
            text = tag.strip()
            if text:
                node = {"tag": "text", "text": text, "children": []}
                stack[-1]["children"].append(node)
        
        elif tag.name == "table":
            # 表格节点
            node = {"tag": "table", "text": str(tag), "children": []}
            stack[-1]["children"].append(node)


        # elif tag.name and tag.get_text().strip():  # 非层级节点，有内容
        #     node = {
        #         "tag": tag.name,
        #         "text": tag.get_text().strip(),
        #         "html": str(tag),
        #         "children": []
        #     }
        #     # 插入到最近的层级节点（栈顶）
        #     stack[-1]["children"].append(node)

    return root

# 示例 HTML
html = """
<body>
<header>
  <h1>Document Title</h1>
  <p>A short description of the document.</p>
  <nav>
    <ul>
      <li>Home</li>
      <li>About</li>
      <li>Contact</li>
    </ul>
  </nav>
</header>
<main>
  <section>
    <h2>Introduction</h2>
    <p>Welcome to the document. This is the introduction section.</p>
  </section>
  <section>
    <h2>Details</h2>
    <p>Further details about the topic.</p>
    <h3>Subsection</h3>
    <p>Details in a subsection.</p>
  </section>
  <article>
    <h3>Another Subsection</h3>
    <ul>
      <li>First point</li>
      <li>Second point</li>
    </ul>
  </article>
  <footer>
    <p>Footer content goes here.</p>
  </footer>
</main>
</body>
"""

# 解析并生成文档树
document_tree = build_document_tree(html)

# 打印结构化文档树
import json
print(json.dumps(document_tree, indent=4, ensure_ascii=False))
```

```python
from bs4 import BeautifulSoup

def build_document_tree(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')

    # 根节点
    root = {"tag": "root", "children": []}
    current_level_nodes = {0: root}  # 按层级追踪当前节点，0 代表根节点

    # 已解析节点集合，存储已处理的节点引用
    parsed_nodes = set()

    # 遍历所有标签
    for tag in soup.descendants:
        # 跳过已解析节点
        if tag in parsed_nodes:
            continue

        if tag.name in ["h1", "h2", "h3", "h4", "h5", "h6"]:  # 标题节点
            level = int(tag.name[1])  # 标题层级
            node = {"tag": tag.name, "text": tag.get_text(), "children": []}

            # 找到最近的有效父节点（比当前层级低一层）
            parent_level = level - 1
            while parent_level not in current_level_nodes:
                parent_level -= 1  # 找到最近的有效父层级
            current_level_nodes[parent_level]["children"].append(node)

            # 更新当前层级的节点
            current_level_nodes[level] = node

            # 标记为已解析
            parsed_nodes.add(tag)

        elif tag.name in ["ul", "ol"]:  # 列表节点
            # 找到最近的层级节点
            highest_level = max(current_level_nodes.keys())
            parent_node = current_level_nodes[highest_level]

            # 创建列表节点
            list_node = {"tag": tag.name, "text": None, "children": []}
            parent_node["children"].append(list_node)

            # 递归解析列表项
            for child in tag.find_all("li", recursive=False):
                if child not in parsed_nodes:
                    list_item = build_list_item_node(child, parsed_nodes)
                    if list_item:
                        list_node["children"].append(list_item)

            # 标记为已解析
            parsed_nodes.add(tag)

        elif tag.name and tag.get_text().strip():  # 非层级节点
            # 找到最近的层级节点
            highest_level = max(current_level_nodes.keys())
            parent_node = current_level_nodes[highest_level]

            # 创建内容节点
            node = {
                "tag": tag.name,
                "text": tag.get_text().strip(),
                "html": str(tag),
                "children": []
            }
            parent_node["children"].append(node)

            # 标记为已解析
            parsed_nodes.add(tag)

    return root


def build_list_item_node(list_item, parsed_nodes):
    """解析列表项及其嵌套内容"""
    node = {
        "tag": "li",
        "text": list_item.get_text(strip=True),
        "children": []
    }

    # 检查列表项是否包含嵌套列表
    for child in list_item.find_all(recursive=False):
        if child in parsed_nodes:
            continue

        if child.name in ["ul", "ol"]:  # 嵌套列表
            nested_list = {"tag": child.name, "text": None, "children": []}
            for nested_item in child.find_all("li", recursive=False):
                if nested_item not in parsed_nodes:
                    nested_list_item = build_list_item_node(nested_item, parsed_nodes)
                    if nested_list_item:
                        nested_list["children"].append(nested_list_item)
            node["children"].append(nested_list)
            parsed_nodes.add(child)
        elif child.name and child.get_text(strip=True):  # 其他内容
            child_node = {
                "tag": child.name,
                "text": child.get_text(strip=True),
                "children": []
            }
            node["children"].append(child_node)
            parsed_nodes.add(child)

    parsed_nodes.add(list_item)  # 标记列表项为已解析
    return node

# 示例 HTML
html = """
<body>
<h1>Document Title</h1>
<p>A short description of the document.</p>
<nav>
  <ul>
    <li>Home</li>
    <li>About
      <ul>
        <li>Team</li>
        <li>History</li>
      </ul>
    </li>
    <li>Contact</li>
  </ul>
</nav>
<h2>Introduction</h2>
<p>Welcome to the document. This is the introduction section.</p>
<section>
  <h3>Subsection</h3>
  <ul>
    <li>Item 1
      <ol>
        <li>Item 1.1</li>
        <li>Item 1.2</li>
      </ol>
    </li>
    <li>Item 2</li>
  </ul>
</section>
<h2>Another Section</h2>
<p>More content here.</p>
</body>
"""

# 解析并生成文档树
document_tree = build_document_tree(html)

# 打印结构化文档树
import json
print(json.dumps(document_tree, indent=4, ensure_ascii=False))
```
